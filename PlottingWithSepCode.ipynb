```python
import physplot as phys
import numpy as np
from __future__ import division

array1 = np.array([np.tan(-0.175),np.tan(-0.345),np.tan(-.524),np.tan(-.689),np.tan(-.873),np.tan(0.175),np.tan(0.345),np.tan(.524),np.tan(.689),np.tan(.873)])
array2 = np.array([5.27,10.22,17.44,25.9,37.0,-5.28,-10.66,-17.4,-27.1,-39.0])
errarray2 = np.array([.2,.2,.2,.2,.2,.2,.2,.2,.2,.2])

#plotfit takes in parameters in this order
#phys.plotfit(x,y,dy,title,'xlabel(units)','ylabel(units)','slope units')  Slope units will not render in LaTeX

phys.plotfit(array1,array2,errarray2,'N=100, Delta Theta vs Current','Angle(radians)','Current(mA)','')



```
